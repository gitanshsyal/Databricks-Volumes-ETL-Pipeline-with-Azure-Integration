```python
valid_df=spark.read.format('delta').load("/Volumes/catalog1/pipeline_db/volumes/silver/data").where("column_category = 'Valid'")
display(valid_df)
                                                                                                     
                                                                                                     
```

```python
dbServer = 'db-server1'
dbPort = '1433'
dbName = 'dbname1'
databricksScope = 'project-scope'
dbUser = dbutils.secrets.get(scope = databricksScope, key=
'username')
dbPassword = dbutils.secrets.get(scope = databricksScope, key=
'password')
connectionUrl ='jdbc:sqlserver://{}.database.windows.net:{};database={};user={};'.format(dbServer,dbPort, dbName, dbUser,dbPassword)
connectionProperties = {
'password': dbPassword,
'driver':'com.microsoft.sqlserver.jdbc.SQLServerDriver'
}
valid_customers_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_customers',
properties = connectionProperties )
valid_store_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_stores',
properties = connectionProperties )
valid_product_df = spark.read.jdbc(url = connectionUrl, table =
'dbo.valid_products',
properties = connectionProperties )
```

```python
modified_df=valid_df.join(valid_product_df,valid_df.product_id==valid_product_df.product_id,how='inner').drop(valid_product_df.product_id)
modified_df=modified_df.join(valid_customers_df,modified_df.customer_id==valid_customers_df.customer_id,how='inner').drop(valid_customers_df.customer_id)
modified_df=modified_df.join(valid_store_df,modified_df.store_id==valid_store_df.stored_id,how='inner').drop(valid_store_df.stored_id)
```

```python
%fs
mkdirs /Volumes/catalog1/pipeline_db/volumes/silver/processed
```

```python
from pyspark.sql.functions import *
modified_df=modified_df.withColumn('NetAmount',col('quantity')*col('price')*(1-col('discount')))
modified_df=modified_df.withColumn('Margin',col('NetAmount')-(col('quantity')*col('product_price')))
modified_df.write.mode('append').format('delta').save('/Volumes/catalog1/pipeline_db/volumes/silver/processed')
```
