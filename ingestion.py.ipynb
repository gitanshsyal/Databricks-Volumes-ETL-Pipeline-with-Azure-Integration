```python
source_schema='transaction_id string,store_id string,customer_id string,product_id string,quantity int,price double,discount double,transaction_date string'
source_df=spark.read.format('csv').option('header',True).schema(source_schema).load('abfss://data@storgeaccount1222.dfs.core.windows.net/catalog/raw')
source_df.createOrReplaceTempView('source_v')
display(source_df)
```

```python
from delta.tables import DeltaTable

bronze_path = "/Volumes/catalog1/pipeline_db/volumes/bronze_data"
if not DeltaTable.isDeltaTable(spark, bronze_path):
  spark.sql('''create table if not exists DELTA.`/Volumes/catalog1/pipeline_db/volumes/bronze_data`
                select * from source_v;
            
            ''')
else:
    spark.sql(f"""
merge into delta.`/Volumes/catalog1/pipeline_db/volumes/bronze_data` t
using (SELECT * FROM source_v) s
on t.transaction_id=s.transaction_id
when matched then
update set t.store_id=s.store_id,t.customer_id=s.customer_id,t.product_id=s.product_id,t.quantity=s.quantity,t.price=s.price,t.discount=s.discount
when not matched then
insert *
    """)
```
